```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../../')
import do_mpc
import pandas as pd
import os
```

```python
%matplotlib notebook
```

```python
file_list = os.listdir('./results/')
n_res = len(file_list)
```

```python
res = []
for file_i in file_list:
    res.append(do_mpc.data.load_results('./results/'+file_i))
    
```

# Theta-phi Darstellung:

```python
n_row = int(np.ceil(np.sqrt(n_res)))
n_col = int(np.ceil(n_res/n_row))

if False:
    fig, ax = plt.subplots(n_row, n_col, sharex=True, sharey=True, figsize=(8,8))
    for i, res_i in enumerate(res):
        ax[i//n_col, np.mod(i,n_col)].plot(res_i['simulator']['_x', 'phi'],res_i['simulator']['_x', 'theta'])
        #ax[i//n_col, np.mod(i,n_col)].axis('off')
else:
    fig, ax = plt.subplots()
    for res_i in res:
        ax.plot(res_i['simulator']['_x', 'phi'],res_i['simulator']['_x', 'theta'])
    
fig.tight_layout(pad=0)
```

# WindstÃ¤rke Darstellung:

```python
fig, ax = plt.subplots(n_row, n_col, sharex=True, sharey=True, figsize=(8,8))
for i, res_i in enumerate(res):
    ax[i//n_col, np.mod(i,n_col)].plot(res_i['simulator']['_time'],res_i['simulator']['_p', 'v_0'])
    
fig.tight_layout(pad=0.1)
```
