```python
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append('../../')
import do_mpc
import pandas as pd
import os
```

```python
file_list = os.listdir('./results/')
n_res = len(file_list)
```

```python
res = []
for file_i in file_list:
    res.append(do_mpc.data.load_results('./results/'+file_i))
```

# Theta-phi Darstellung:

```python
n_row = int(np.ceil(np.sqrt(n_res)))
n_col = int(np.ceil(n_res/n_row))

fig, ax = plt.subplots(n_row, n_col, sharex=True, sharey=True, figsize=(14,14))

for i, res_i in enumerate(res):
    ax[i//n_col, np.mod(i,n_col)].plot(res_i['simulator']['_x', 'phi'],res_i['simulator']['_x', 'theta'])
    
fig.tight_layout(pad=0.1)
```

# Windstärke Darstellung:

```python
fig, ax = plt.subplots(n_row, n_col, sharex=True, sharey=True, figsize=(14,14))
for i, res_i in enumerate(res):
    ax[i//n_col, np.mod(i,n_col)].plot(res_i['simulator']['_time'],res_i['simulator']['_p', 'v_0'])
    
fig.tight_layout(pad=0.1)
```

# E_0 Darstellung:

```python
fig, ax = plt.subplots(n_row, n_col, sharex=True, sharey=True, figsize=(14,14))
for i, res_i in enumerate(res):
    ax[i//n_col, np.mod(i,n_col)].plot(res_i['simulator']['_time'],res_i['simulator']['_p', 'E_0'])
    
fig.tight_layout(pad=0.1)
```
